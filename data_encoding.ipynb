# handling categorical data :
machine learning algorithm can read only in numeric data.the common technique are a) label encoding as a simple technique of assigning a unique integer value to each category in critical variable

```python
import numpy  as np
from sklearn.preprocessing import LabelEncoder
```

```python
gender= np.array(['male','female','female','male','thirdgen'])
```

```python
le=LabelEncoder()
```

```python
gender_encoder=le.fit_transform(gender)
```

```python
gender_encoder
```

for lael encoding in python, LabelEncoder() object is required from sklearn.preprocessing module.the code above show that,gender is categorical variable which is encoded with lael encoder(),which asign 1 to male,0 to female.

b) one-hot encoding on the oter hand ,hot encoding creates a column for each category in a catehorical variable
